# Training the model

```python
from sklearn.model_selection import train_test_split
rawReviews = df['REVIEW']
reviews = df['REVIEW_punctuation_tokenized_sw_lemmatized'].values
labels = df['SENTIMENT'].values
reviews_train, reviews_test, y_train, y_test = train_test_split(reviews, labels, test_size=0.45, random_state=1000)
```

## Transforming The Text into Numerical Feature Vectors

```python

from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English
import string
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer( ngram_range=(1,2))
vectorizer.fit(reviews_train)
x_train = vectorizer.transform(reviews_train)
x_test = vectorizer.transform(reviews_test)
print(x_test)
```

## To Visualize the vectorize Matrix

```python
df_sample = df[0:15]
cv2 = CountVectorizer(ngram_range=(1,2))
#The bellow code returns an error, (why?) thats why I replaced it with another code
# x = cv2.fit_transform(df_sample['REVIEW_clean_tokenized_no_sw_lemmatized'])
x_sample = cv2.fit_transform(df_sample['REVIEW_punctuation_tokenized_sw_lemmatized'])

print(x_sample.shape)
dd = pd.DataFrame(x_sample.toarray(), columns=cv2.get_feature_names_out())
dd.head(20)
```

## Continuing The Training 

```python

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

## Evaluation and Accuracy

```python
accuracy = classifier.score(x_test, y_test)
print('Accuracy:', accuracy)
```

# Predictions on New Data

```python
new_reviews = ['garbage', 'have some errors, but not five stars', 'Clear and concise']
X_new = vectorizer.transform(new_reviews)
classifier.predict(X_new)
```
